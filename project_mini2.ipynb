```python

from pandas import * 
file = 'AppleStore.csv'         
f = read_csv(file)
ra = []
for l in f.values:
    if l[-5] == 'Games':
        ra.append(l[8])
#print(ra)
ral = len(ra)
from functools import *
totr = reduce(lambda x,y:x+y,ra)
avgrG = totr/ral
print('average rating of gaming app:',avgrG)

nra = []
for li in f.values:
    if li[-5] != 'Games':
        nra.append(li[8])
nral = len(nra)
#print(nra)
totnr = reduce(lambda x,y:x+y,ra)
avgrnG = totnr/nral
print('average rating of non gaming app:',avgrnG)

##
lis = []
n=0
for i in f.values:
    if i[5] == 0:
        n+=1
lis.append(n)
for i in f.values:
    if i[5]<20 and i[5]!=0:
        n+=1
lis.append(n)
for i in f.values:
    if i[5]<50 and i[5]>20:
        n+=1
lis.append(n)
for i in f.values:
    if i[5]>50:
        n+=1
lis.append(n)
#print(lis)
from matplotlib.pyplot import *
lab =['Free apps','Affordable apps','Expensive apps','Very expensive']
pie(lis,labels = lab,autopct = '%1.2f%%')
legend(loc = 'best')
show()
```
